```python
import scipy.io as sio
import numpy as np
from sklearn import svm
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
MNIST = sio.loadmat('mnist_data.mat')
SPAM  = sio.loadmat('spam_data.mat')
CIFAR_10 = sio.loadmat('cifar10_data.mat')

SEED  = 42
np.random.seed(SEED)
```

```python
########## Problem 2 ################################################
# A
mnist_train, mnist_t_labels = MNIST['training_data'], MNIST['training_labels']
mnist = list(zip(mnist_train, mnist_t_labels))
np.random.shuffle(mnist)
mnist_val, mnist_train = mnist[:10000], mnist[10000:]
```

```python
#B
spam_train, spam_t_labels = SPAM['training_data'], SPAM['training_labels']
spam = list(zip(spam_train, spam_t_labels))
np.random.shuffle(spam)
spam_val, spam_train = spam[: int(len(spam) * 0.2)], spam[int(len(spam) * 0.2):]
```

```python
#C
c_10_train, c_10_t_labels = CIFAR_10['training_data'], CIFAR_10['training_labels']
c_10 = list(zip(c_10_train, c_10_t_labels))
np.random.shuffle(c_10)
c_10_val, c_10_train = c_10[: 5000], c_10[5000:]
```

```python
######## Problem 3 ##################################################
#A
mnist_t_size = [100, 200, 500, 1000, 2000, 5000, 10000]
mnist_train_acc = []
mnist_val_acc = []
for size in mnist_t_size:
    data, labels = [x for x,y in mnist_train[:size]], [y for x,y in mnist_train[:size]]
    val_data, val_labels = [x for x,y in mnist_val], [y for x,y in mnist_val]
    lin_clf = svm.LinearSVC()
    lin_clf.fit(data, np.ravel(labels))
    mnist_train_acc.append(accuracy_score(labels, lin_clf.predict(data)))
    mnist_val_acc.append(accuracy_score(lin_clf.predict(val_data), val_labels))

plt.plot(mnist_t_size, mnist_train_acc, label = 'Training Accuracy')
plt.plot(mnist_t_size, mnist_val_acc, label = 'Validation Accuracy')
plt.legend()
plt.title('Training Accuracy of MNIST Dataset')
plt.xlabel('Size of Training Set')
plt.ylabel('Accuracy')
```

```python
#B
spam_t_size = [100, 200, 500, 1000, 2000, len(spam_train)]
spam_train_acc = []
spam_val_acc = []
for size in spam_t_size:
    data, labels = [x for x,y in spam_train[:size]], [y for x,y in spam_train[:size]]
    val_data, val_labels = [x for x,y in spam_val], [y for x,y in spam_val]
    lin_clf = svm.LinearSVC()
    lin_clf.fit(data, np.ravel(labels))
    spam_train_acc.append(accuracy_score(labels, lin_clf.predict(data)))
    spam_val_acc.append(accuracy_score(val_labels, lin_clf.predict(val_data)))

plt.plot(spam_t_size, spam_train_acc, label = 'Training Accuracy')
plt.plot(spam_t_size, spam_val_acc, label = 'Validation Accuracy')
plt.legend()
plt.title('Training Accuracy of SPAM Dataset')
plt.xlabel('Size of Training Set')
plt.ylabel('Accuracy')
```

```python
c_10_t_size = [100, 200, 500, 1000, 2000, 5000]
c_10_train_acc = []
c_10_val_acc = []
for size in c_10_t_size:
    data, labels = [x for x,y in c_10_train[:size]], [y for x,y in c_10_train[:size]]
    val_data, val_labels = [x for x,y in c_10_val], [y for x,y in c_10_val]
    lin_clf = svm.LinearSVC()
    lin_clf.fit(data, np.ravel(labels))
    c_10_train_acc.append(accuracy_score(labels, lin_clf.predict(data)))
    c_10_val_acc.append(accuracy_score(val_labels, lin_clf.predict(val_data)))

plt.plot(c_10_t_size, c_10_train_acc, label = 'Training Accuracy')
plt.plot(c_10_t_size, c_10_val_acc, label = 'Validation Accuracy')
plt.legend()
plt.title('Training Accuracy of CIFAR-10 Dataset')
plt.xlabel('Size of Training Set')
plt.ylabel('Accuracy')
```

```python
####### Problem 4 ############################################################
c_vals = [10 ** x for x in range (-10, 10)]
acc = []
for c in c_vals:
    data, labels = [x for x,y in mnist_train], [y for x,y in mnist_train]
    val_data, val_labels= [x for x,y in mnist_val], [y for x,y in mnist_val]
    lin_clf = svm.LinearSVC(C = c)
    lin_clf.fit(data, np.ravel(labels))
    acc.append(accuracy_score(val_labels, lin_clf.predict(val_data)))

```

```python
hyper_p = list(zip(c_vals, acc))
optimal_c = max(hyper_p, key = lambda tup: tup[1])[0]
hyper_p, "hyperparameter values and corresponding accuracy", optimal_c, "optimal C"
```

```python
###### Problem 5 ###################################################
np.random.shuffle(spam)
cross_val = []
c_vals = [10 ** x for x in range (-10, 10)]
acc = []
k = 5
size = int(len(spam) / k)

for i in range(k):
    t_set = spam[i * size: (i + 1) * size]
    cross_val.append(t_set)
    
for c in c_vals:
    k_acc = []
    for i in range(k):
        train_set = []
        val_set = []
        for j in range(k):
            if j == i:
                val_set += cross_val[j]
            else:
                train_set += cross_val[j]
        data, labels = [x for x,y in train_set], [y for x,y in train_set]
        lin_clf = svm.LinearSVC(C = c)
        lin_clf.fit(data, np.ravel(labels))
        val_data, val_labels = [x for x,y in val_set], [y for x,y in val_set]
        k_acc.append(accuracy_score(val_labels, lin_clf.predict(val_data)))
    acc.append(sum(k_acc) / k)
```

```python
hyper_p = list(zip(c_vals, acc))
optimal_c = max(hyper_p, key = lambda tup: tup[1])[0]
hyper_p, "hyperparameter values and corresponding accuracy", optimal_c, "optimal C"
```

```python
# Checks kaggle submission for formatting correctness. 
# Usage: python check.py <competiton_name, eg. mnist> <submission csv file>

import argparse
import pandas as pd

competitions = ['mnist', 'spam', 'cifar10']
config = {
    'mnist': 
            {
             'rows': 10000,
             'range': (0, 9),
             },
    'spam': 
            {'rows': 5857,
             'range': (0, 1),
             },
    'cifar10': 
            {'rows': 10000,
             'range': (0, 9),
             }
}


parser = argparse.ArgumentParser()
parser.add_argument("comp_name",
                        choices=competitions, 
                        help="Name of competition you want to check against")
parser.add_argument("fname", 
                        help="Filename")
args = parser.parse_args()


print("Competition chosen: {}".format(args.comp_name))
print("File chosen: {}".format(args.fname))

# Load data
# print("Loading csv")
df = pd.read_csv(args.fname)


# Check the column names
print("Checking column names...")
if ['Id', 'Category'] != list(df.columns.values):
    print("Column header name check failed; most likely a spelling mistake OR" + \
            " a space was put between 'Id,' and 'Category'")
    exit(1)

# Check the row count
# print("Checking row count...")
rows_count = df.shape[0]
if rows_count != config[args.comp_name]['rows']:
    print("Row count should be {} when it is {}".format(config[args.comp_name]['rows'], rows_count))
    exit(1)

# Check the indexes
if list(range(1, rows_count+1)) != list(df['Id']):
    print("Range of Ids is wrong. Most likely submission is 0-indexed instead of 1-indexed")
    exit(1)

# Check the range of the category and the Ids. 
low, hi = config[args.comp_name]['range']
if not df['Category'].between(low, hi).all(): 
    print("Range of Category is wrong")
    exit(1)

# Check the dtype
if df['Id'].dtype != 'int64':
    print("Id is not of type int64, instead is of type {}".format(df['Id'].dtype))
    exit(1)

if df['Category'].dtype != 'int64':
    print("Category is not of type int64")
    exit(1)

print("All basic sanity checks passed!")




```
