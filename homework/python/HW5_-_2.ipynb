# 2.1, 2.2 Decision Tree / Random Forest Code

```python
class DecisionTree:    
    class tree_node:
        def __init__(self, decision_rule, left, right, label):
            self.decision_rule = decision_rule
            self.left = left
            self.right = right
            self.label = label
            self.id = str(uuid.uuid4())
        
        def __repr__(self):
            if self.decision_rule:
                idx, thresh = self.decision_rule
                return str(t_features[idx]) + "  <=  "  + str(thresh)
            else:
                
                return t_names[int(self.label)]
        
     
    def __init__(self, max_depth = 5, purity_thresh = 0.90, feat_thresh = 0.30, split_thresh = 0.95):
        """
        TODO: initialization of a decision tree
        """
        self.max_depth = max_depth
        self.purity = purity_thresh
        self.feat_thresh = feat_thresh
        self.split_thresh = split_thresh
        
    
    @staticmethod
    
    
    def entropy(y):
        """
        TODO: implement a method that calculates the entropy given all the labels
        """
        counter = Counter(y)
        return scipy.stats.entropy(list(counter.values()))
        
        

    @staticmethod
    def information_gain(X, y, thresh, idx):
        """
        TODO: implement a method that calculates information gain given a vector of features
        and a split threshold
        """
        ##RETURNS THE FEATURE
        entropy_before = DecisionTree.entropy(y)
        l_x, l_y, r_x, r_y = DecisionTree.split(X, y, idx, thresh)
        e_left, e_right = DecisionTree.entropy(l_y), DecisionTree.entropy(r_y)
        tot_l, tot_r = len(l_y), len(r_y)
        entropy_after = ((tot_l * e_left) + (tot_r * e_right)) / (tot_l + tot_r)
        return entropy_before - entropy_after


    def splits(self, X):
        splits = {}
        s, r = 0, 0
        for i in self.feats:
            splits[i] = np.unique(X[:,i])
            s += len(splits[i])
        
        rand_split = np.random.uniform(0, 1, s)
        for col in self.feats:
            if len(splits[col]) <= 2:
                r += 1
                splits[col] = [splits[col][0]]
            else:
                f_splits = list()
                for s in splits[col]:
                    if rand_split[r] <= self.feat_thresh:
                        f_splits.append(s)
                if len(f_splits) != 0:
                    splits[col] = f_splits
        return splits
    
    @staticmethod
    def split(X, y, idx, thresh):
        """
        TODO: implement a method that return a split of the dataset given an index of the feature and
        a threshold for it
        """
        left = np.where(X[:, idx] <= thresh)[0]
        right = np.where(X[:, idx] > thresh)[0]
        left_X, left_y =  X[left, :], y[left]
        right_X, right_y = X[right, :], y[right]
        return left_X, left_y, right_X, right_y
    
   
    
    def feats(self, X):
        rand_cols = np.random.uniform(0,1, X.shape[1])
        feats = []
        rand_idx = 0
        for i in range(X.shape[1]):
            if rand_cols[rand_idx] <= self.feat_thresh:
                feats.append(i)
            rand_idx += 1
        
        return feats
                
    def segmenter(self, X, y):
        """
        TODO: compute entropy gain for all single-dimension splits,
        return the feature and the threshold for the split that
        has maximum gain
        """
        #RETURN THE BEST SPLIT
        features_and_thresholds = []
        
        for idx in self.splits.keys():
            for thresh in self.splits[idx]:
                f_t_gain = (idx, thresh, DecisionTree.information_gain(X, y, thresh, idx))
                features_and_thresholds.append(f_t_gain)
        idx, thresh, entropy = max(features_and_thresholds, key = lambda x: x[2])
        return idx, thresh
    
    def fit(self, X, y):
        """
        TODO: fit the model to a training set. Think about what would be 
        your stopping criteria
        """
        self.feats = self.feats(X)
        self.splits = self.splits(X)
        self.tree = self.grow_tree(X, y, 0)

    
    def is_pure(self,y):
        counter = Counter(y)
        for label in counter.keys():
            if counter[label] / sum(counter.values()) > self.purity:
                return True
        return False
        
    
    def leaf_label(self, y):
        c = Counter(y)
        return c.most_common(1)[0][0]
    
    
    def grow_tree(self, X, y, depth):
        if y is None or len(y) == 0:
            return
        if self.is_pure(y) or depth  == self.max_depth:
            label = self.leaf_label(y)
            return DecisionTree.tree_node(None, None, None, label)   
        else:
            idx, thresh = self.segmenter(X, y)
            x_l, y_l, x_r, y_r = DecisionTree.split(X, y, idx, thresh)  
            left = self.grow_tree(x_l, y_l, depth + 1)
            right = self.grow_tree(x_r, y_r, depth + 1)
            return DecisionTree.tree_node((idx, thresh), left, right, None)

    def predict(self, X):
        """
        TODO: predict the labels for input data 
        """
        num_points = len(X)
        labels = []
        categorical_data = ['sex', 'pclass', 'ticket', 'cabin', 'embarked']
        
        for i in range(num_points):
            point = X[i,:]
            tree_helper = self.tree
            while tree_helper.left is not None and tree_helper.right is not None:
                if tree_helper.label:
                    labels.append(tree_helper.label)
                    continue
                feature, thresh = tree_helper.decision_rule
                if point[feature] <= thresh:
                    tree_helper = tree_helper.left
                else:
                    tree_helper = tree_helper.right
            labels.append(tree_helper.label)
        return labels
    
    def pred_one(self, X):
        """
        TODO: predict the labels for input data 
        """
        tree_helper = self.tree
        while tree_helper.label is None:
            idx, thresh = tree_helper.decision_rule
            print(features[idx], "<=", thresh, ":This is my Decision Rule")
            print(X[idx], ':This is the pts. Feature Value')
            if X[idx] <= thresh:
                tree_helper = tree_helper.left

            else:
                tree_helper = tree_helper.right
        print('Therefore this is the label: ', class_names[int(tree_helper.label)])
        

    def __repr__(self):
        """
        TODO: one way to visualize the decision tree is to write out a __repr__ method
        that returns the string representation of a tree. Think about how to visualize 
        a tree structure. You might have seen this before in CS61A.
        """
        return str(self.max_depth) + ', this is the max_depth ' +  str(self.purity) + ', this is my purity thresh'

class RandomForest():
    def __init__(self, max_depth = 2, num_trees = 3, purity_thresh = 0.90, feat_thresh = 0.75, split_thresh = 0.75):
        """
        TODO: initialization of a random forest
        """
        self.max_depth = max_depth
        self.num_trees = num_trees
        self.purity_thresh = purity_thresh
        self.feat_thresh = feat_thresh
        self.split_thresh = split_thresh
        
    def __repr__(self):
        return 'im a random forest skrrrt'

    def fit(self, X, y):
        """
        TODO: fit the model to a training set.
        """
        m_d, p_t, f_t, s_t = self.max_depth, self.purity_thresh, self.feat_thresh, self.split_thresh
        self.trees = [DecisionTree(m_d, p_t, f_t, s_t) for _ in range(self.num_trees)]
        i = 0
        for tree in self.trees:
            tree.fit(X,y)
            print('tree was fit', str(i))
            i += 1
    
    def predict(self, X):
        votes, pred = [], []
        for tree in self.trees:
            votes.append(tree.predict(X))
            
        votes = np.array(votes)
        for i in range(len(X)):
            votes_i = Counter(votes[:,i])
            winner = votes_i.most_common()[0][0]
            pred.append(winner)
        return pred
    
 
```

```python
def visualize(tree):
    q = Queue()
    dot = Digraph(comment = 'Titanic Decision Tree Visualization')
    t = tree
    q.put((None, t))
    while not q.empty():
        parent, child = q.get()
        if not parent:
            dot.node(child.id, repr(child))
        else:
            dot.node(child.id, repr(child))
            dot.edge(parent.id, child.id, constraint = 'false')
        for g in [child.right, child.left]:
            if g:
                q.put((child, g))
        
    dot.render('titanic_tree.pdf', view = True)
    
```
